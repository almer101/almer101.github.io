```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
plt.style.use('bmh')
```

```python
N = 1000
B = 1
dt = 1/N
sigma = 1

w = [0]

while w[-1] < B:
    dw = sigma * np.sqrt(dt) * np.random.normal(0,1)
    w.append(w[-1] + dw)

ts = np.linspace(0, dt * (len(w)-1), len(w))
plt.plot(ts, w)
```

```python
def simulate_BM(dt, B):
    max_iter = 10000
    w = [0]
    while w[-1] < B:
        dw = sigma * np.sqrt(dt) * np.random.normal(0,1)
        w.append(w[-1] + dw)
        if len(w) > max_iter:
            return ([0],[0])

    ts = np.linspace(0, dt * (len(w)-1), len(w))
    return (ts, w)
```

```python
max_t = 0
sum_t = 0
cnt = 0
for i in tqdm(range(1000)):
    ts, w = simulate_BM(dt, B=B)
    if len(ts) == 1: continue
    sum_t += ts[-1]
    cnt += 1
    # plt.plot(ts,w)
    max_t = max(max_t, ts[-1])

# plt.plot([0,max_t+1], [B, B], label=f'B={B}')
# plt.xlim((0,max_t+1))
# plt.legend()

print(sum_t/cnt)
```
